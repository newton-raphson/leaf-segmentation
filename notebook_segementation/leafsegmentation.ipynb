```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Read the image
image = cv2.imread('2.jpg')  # Replace 'your_image.jpg' with your image file path

# Define the coordinates of the region you want to crop (top-left and bottom-right)
# Crop the image as per the requirement
image = image[0:898, 0:750]
# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(image, cmap = 'gray', interpolation = 'bicubic')
```

THRESHOLDING THE IMAGES TO REMOVE HIGH INTENSITY PIXELS

```python
# Set a threshold value to distinguish high and low intensities
threshold_value = 80  # Adjust as needed(80 for image 2, 30 for image 1)

# Apply thresholding to create a binary mask
_, binary_mask = cv2.threshold(gray, threshold_value, 255, cv2.THRESH_BINARY)

# Bitwise AND the original image with the binary mask
thresholded = cv2.bitwise_and(image, image, mask=binary_mask)
thresholded
```

REMOVING HORIZONTAL LINES

```python
# Create a horizontal kernel for morphological operations
kernel = np.ones((1, 10), np.uint8)

# Use morphology to remove horizontal lines from the thresholded image
lines_removed = cv2.morphologyEx(thresholded, cv2.MORPH_OPEN, kernel)
lines_removed
```

REMOVE VERTICAL LINES


```python
# Create a horizontal kernel for morphological operations
kernel_vert = np.ones((25, 1), np.uint8)

# Use morphology to remove horizontal lines from the thresholded image
lines_removed_vertical = cv2.morphologyEx(thresholded, cv2.MORPH_OPEN, kernel_vert)
lines_removed_vertical
```

```python
# Subtract lines_removed from thresholded to get the lines only
lines_remove_horizontal = cv2.subtract(thresholded, lines_removed)
lines_removed_both = cv2.subtract(lines_remove_horizontal, lines_removed_vertical)

lines_removed_both
```

BLURRED IMAGE TO REMOVE NOISE


```python
# Convert the image to grayscale
gray = cv2.cvtColor(lines_removed_both, cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur to reduce noise
blurred_img = cv2.GaussianBlur(gray, (3, 3), 0)
blurred_img
```

HISTOGRAM EQUALIZATION

```python

equalized_img = cv2.equalizeHist(blurred_img)
equalized_img
```

SATURATION INCREMENT


```python
# Increase saturation for color enhancement
hsv_img = cv2.cvtColor(lines_removed_both, cv2.COLOR_BGR2HSV)
hsv_img[:, :, 1] = hsv_img[:, :, 1] * 2.5  # Increase saturation factor
saturated_img = cv2.cvtColor(hsv_img, cv2.COLOR_HSV2BGR)
saturated_img
```
